````python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-generation", model="Qwen/Qwen2.5-Coder-1.5B")
messages = [
    {"role": "user", "content": "Who are you?"},
]
pipe(messages)
````

````python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-Coder-1.5B")
model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen2.5-Coder-1.5B")
messages = [
    {"role": "user", "content": "Who are you?"},
]
inputs = tokenizer.apply_chat_template(
	messages,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to(model.device)

outputs = model.generate(**inputs, max_new_tokens=40)
print(tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
````

## Step 4: Load Model and Tokenizer with Unsloth

````python
from unsloth import FastLanguageModel
import torch

# Define model and tokenizer parameters
max_seq_length = 2048 # Max sequence length for your model
dtype = None # None for auto detection. Choose from torch.float16, torch.bfloat16, or torch.float32
load_in_4bit = True # Use 4bit quantization to reduce memory usage

# Load the model and tokenizer
print("Loading Qwen/Qwen2.5-Coder-1.5B model and tokenizer...")
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "Qwen/Qwen2.5-Coder-1.5B",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)

print("MODEL LOADED")
````

## Step 5: Attach LoRA Adapter

````python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16,
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
    target_modules = [
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj",
        "gate_proj",
        "up_proj",
        "down_proj",
    ],
    random_state = 3407,
    max_seq_length = max_seq_length,
)

print("LoRA adapter attached. Trainable parameters:")
model.print_trainable_parameters()
````

## Step 6: Format Dataset for Training

````python
from datasets import DatasetDict, load_dataset

# Load the datasets to ensure they are defined in the current scope
train_dataset = load_dataset('json', data_files='data/train.jsonl', split='train')
val_dataset = load_dataset('json', data_files='data/val.jsonl', split='train')

# Define the formatting function to handle batches
def formatting_func(examples):
    instructions = examples["instruction"]
    inputs_text = examples["input"]
    outputs_text = examples["output"]

    formatted_texts = []
    for instruction, input_text, output_text in zip(instructions, inputs_text, outputs_text):
        # The format required is: Instruction: {instruction}\n\nInput: {input}\n\nOutput: {output}
        # Handle cases where 'input' might be empty
        if input_text:
            text = f"Instruction:\n{instruction}\n\nInput:\n{input_text}\n\nOutput:\n{output_text}"
        else:
            text = f"Instruction:\n{instruction}\n\nOutput:\n{output_text}"
        formatted_texts.append(text)

    return { "text" : formatted_texts }

# Apply the formatting function to both datasets
print("Applying formatting function to training dataset...")
formatted_train_dataset = train_dataset.map(formatting_func, remove_columns=['instruction', 'input', 'output'], batched=True)
print("Applying formatting function to validation dataset...")
formatted_val_dataset = val_dataset.map(formatting_func, remove_columns=['instruction', 'input', 'output'], batched=True)

# Create a DatasetDict
datasets = DatasetDict({
    'train': formatted_train_dataset,
    'validation': formatted_val_dataset
})

print("\nDatasets after formatting:")
print(datasets)

print(f"\nTrain dataset rows: {len(datasets['train'])}")
print(f"Validation dataset rows: {len(datasets['validation'])}")

print("\nFirst formatted sample from the training dataset:")
print(datasets["train"][0]["text"])
````

## Step 7: Create `SFTTrainer`

````python
from trl import SFTTrainer
from transformers import TrainingArguments

# Define the training arguments
training_arguments = TrainingArguments(
    per_device_train_batch_size = 2,
    gradient_accumulation_steps = 4,
    warmup_steps = 20,
    num_train_epochs = 3,
    learning_rate = 2e-4,
    fp16 = True, # Use fp16 for Tesla T4
    logging_steps = 10,
    output_dir = "qwen-coder-lora",
    optim = "paged_adamw_8bit", # Recommended optimizer for LoRA fine-tuning
    seed = 3407,
    report_to = "none",
    save_strategy = "no", # CRITICAL: Disable checkpoint saving to prevent PicklingError
    save_total_limit = 0, # CRITICAL: Ensure no checkpoints are saved
)

# Initialize the SFTTrainer
trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = datasets['train'],
    eval_dataset = datasets['validation'], # Use the validation split for evaluation
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    args = training_arguments,
)

print("SFTTrainer initialized successfully:")
print(trainer)

print("\nTrainingArguments used:")
print(trainer.args)
````

## Step 8: Start Training (`trainer.train()`)

````python
print("Starting model training...")
trainer.train()
print("Model training completed.")
````

### Explanation for avoiding `PicklingError`:

The `PicklingError: Can't pickle trl.trainer.sft_config.SFTConfig` often occurs when the `SFTTrainer` (which internally uses `transformers.Trainer`) attempts to serialize its state for checkpointing. This serialization process involves pickling various objects, and certain configurations or internal states might not be directly pickleable.

To explicitly prevent this, we've set `save_strategy = "no"` and `save_total_limit = 0` in the `TrainingArguments`. This instructs the trainer **not to save any checkpoints** during or after training, thereby completely bypassing the problematic pickling of the `SFTConfig` or any related objects that might cause the error during serialization. This ensures `trainer.train()` can complete without interruption from checkpointing failures.
